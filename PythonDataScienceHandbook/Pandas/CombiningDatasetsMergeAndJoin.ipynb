# Combining Datasets: merge and join

````python
import numpy as np
import pandas as pd
````

````python
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}{1}
    """
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_()) for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a)) for a in self.args)
````

## Categories of Joins

There are two way we can use merge functionality of pandas:

```python
1. pd.merge(left_df, right_df)

2. left_df.merge(right_df)
```
    
The second one is more convenient and compact for use.

### One-to-One Joins

````python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})

df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})

display('df1', 'df2', 'pd.merge(df1, df2)')
````

### Many-to-One Joins

````python
df4 = pd.DataFrame({'group': ['Accounting', 'Engineering', 'HR'],
                    'supervisor': ['Carly', 'Guido', 'Steve']})
display('df1', 'df4', 'pd.merge(df1, df4)')
````

### Many-to-Many Joins

````python
df5 = pd.DataFrame({'group': ['Accounting', 'Accounting', 'Engineering', 'Engineering', 'HR', 'HR'],
                    'skills': ['math', 'spreadsheets', 'software', 'math', 'spreadsheets', 'organization']})

display('df1', 'df5', 'pd.merge(df1, df5)')
````

## Specification of the Merge Key

### The on Keyword

````python
display('df1', 'df2', "pd.merge(df1, df2, on='employee')")
````

````python
df1.merge(df2)
````

### The left_on and right_on Keywords

````python
df3 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                            'salary': [70000, 80000, 120000, 90000]})

display('df1', 'df3', 'pd.merge(df1, df3, left_on="employee", right_on="name")')
````

````python
df1.merge(df3, left_on="employee", right_on="name")
````

````python
pd.merge(df1, df3, left_on="employee", right_on="name").drop('name', axis='columns')
````

````python
df1.merge(df3, left_on="employee", right_on="name").drop('name', axis='columns')
````

### The left_index and right_index Keywords

````python
df1
````

````python
df1a = df1.set_index("employee")
df2a = df2.set_index("employee")
````

````python
display('df1a', 'df2a', 'pd.merge(df1a, df2a, left_index=True, right_index=True)')
````

For convenience, Pandas includes the ```DataFrame.join()``` method, which performs an index-based merge without extra keywords

````python
df1a.join(df2a)
````

````python
pd.DataFrame.join(df1a, df2a)
````

````python
display('df1a', 'df3', "df1a.merge(df3, left_index=True, right_on='name')")
````

## Specifying Set Arithmetic for Joins

````python
df6 = pd.DataFrame({'name': ['Peter', 'Paul', 'Mary'],
                    'food': ['fish', 'beans', 'bread']},
                    columns=['name', 'food'])

df7 = pd.DataFrame({'name': ['Mary', 'Joseph'],
                    'drink': ['wine', 'beer']},
                    columns=['name', 'drink'])

display('df6', 'df7', 'df6.merge(df7)')
````

````python
pd.merge(df6, df7, how='inner')
````

````python
df6.merge(df7, how='outer')
````

````python
display('df6', 'df7', "df6.merge(df7, how='left')") # contains all left entries
````

````python
display('df6', 'df7', "df6.merge(df7, how='right')") # contains all right entries
````

## Overlapping Column Names: The suffixes Keyword

````python
df8 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'rank': [1, 2, 3, 4]})

df9 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'rank': [3, 1, 4, 2]})

display('df8', 'df9', 'df8.merge(df9, on="name")')
````

````python
display('df8', 'df9', 'df8.merge(df9, on="name", suffixes=["_L", "_R"])')
````

## Example: US States Data

````python
pop = pd.read_csv('data/state-population.csv')
areas = pd.read_csv('data/state-areas.csv')
abbrevs = pd.read_csv('data/state-abbrevs.csv')

display('pop.head()', 'areas.head()', 'abbrevs.head()')
````

````python
merged = pd.merge(pop, abbrevs, how='outer', left_on='state/region', right_on='abbreviation')
merged.head()
````

````python
merged = merged.drop('abbreviation', axis=1)
merged.head()
````

````python
merged.isnull().any()
````

````python
merged.loc[merged['population'].isnull(), ['state/region', 'year'] ]
````

````python
merged.loc[merged['state'].isnull(), 'state/region'].unique()
````

````python
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'

merged.isnull().any()
````

````python
final = pd.merge(merged, areas, on='state', how='left')
final.head()
````

````python
final.isnull().any()
````

````python
final['state'][final['area (sq. mi)'].isnull()].unique()
````

````python
final.dropna(inplace=True) 
final.head()
````

````python
final.isnull().any()
````

````python
data2010 = final.loc[(final['year']==2010) & (final['ages']=='total'), :]
````

````python
data2010.head()
````

````python
data2010.set_index('state', inplace=True)
````

````python
data2010.head()
````

````python
density = data2010['population'] / data2010['area (sq. mi)']
density.sort_values(ascending=False, inplace=True)
density.head()
````
