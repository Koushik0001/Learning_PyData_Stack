# Customizing Plot Legends

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0, 10, 1000)
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x), '-g', label='y = sin(x)')
ax.plot(x, np.cos(x), '-b', label='y = cos(x)')
ax.set(xlabel='x', ylabel='y', ybound=(-1.5, 1.5), xbound=(-0.5, 10.5))
ax.legend();
```
